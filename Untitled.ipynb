```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Datos simulados de llamadas recibidas y atendidas
llamadas_recibidas = np.array([1000, 1110, 1110, 1240, 1100, 1200, 1050, 1030, 1050, 1090, 1130, 1150])
llamadas_atendidas = np.array([1000, 1105, 1110, 1250, 1100, 1200, 1050, 1000, 1020, 1040, 1060, 1070])
```

```python
len(llamadas_recibidas)
```

```python
dates = pd.date_range(start='2022-01-31', periods = 12, freq='ME')
```

```python
dates_f = dates.strftime('%B').str[:3]
```

```python
plt.figure(figsize=(10, 5), dpi=90)
plt.plot(dates_f, llamadas_recibidas, lw = 3, color='yellow', marker ='o')
plt.plot(dates_f, llamadas_atendidas, lw = 3, color='royalblue', marker ='o')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.gca().spines['bottom'].set_color('green')
plt.gca().spines['left'].set_color('green')
plt.gca().tick_params(colors='0.1', grid_color='green')

plt.text(0.97, 0.6, 'Recibidas', transform=plt.gca().transAxes,
        color = 'grey', fontsize = 'x-large', fontweight = 'bold')

plt.text(0.97, 0.3, 'Atendidas', transform=plt.gca().transAxes,
        color = 'grey', fontsize = 'x-large', fontweight = 'bold')

plt.ylabel('Cantidad mensual de llamadas', color='grey')
plt.title('Comparacion de llamdas recibidas vs atendidas', color='grey', size =12)

for i in range(7, len(dates)):
    plt.text(i+0.15, llamadas_recibidas[i]+20, llamadas_recibidas[i], color='grey', fontsize=12, ha='center')
    plt.text(i+0.15, llamadas_atendidas[i]-40, llamadas_atendidas[i], color='royalblue', fontsize=12, ha='center')

plt.ylim(900, 1300)
plt.xlim(0,12)
```
