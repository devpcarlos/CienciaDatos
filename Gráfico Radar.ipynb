```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
converExcel = pd.read_excel('Datos_Shale.xlsx')
data_Shale = converExcel.to_csv('Datos_Shales.csv', index=None, header=True)
data_Shale = pd.read_csv('Datos_Shales.csv')
print(data_Shale)
```

```python
labels = data_Shale['Shale']
data = data_Shale['TOC']
```

```python
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
```

```python
data = np.concatenate((data, [data[0]]))
angles += angles[:1]
```

```python
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, data, color='blue', linewidth=2)
ax.fill(angles, data, color='blue', alpha=0.25)
ax.set_theta_offset(np.pi/2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_rgrids(np.arange(0, 15, 2), angle=0, weight='light')

plt.show()
```
